Q2: What is Cumulative Density Function (CDF)? Explain with an example. Why CDF is used?

Cumulative distribution function is a function that gives the probability of a random variable being less than or equal to a certain value. It is useful for analyzing and understanding the probabilities of different events and ranges of values in a continuous random variable.

Example

Suppose we have a normally distributed continuous random variable, such as the heights of a population. The CDF of the normal distribution gives the probability that a randomly selected individual from the population has a height less than or equal to a given height.

Let's say that the mean height of the population is 68 inches, and the standard deviation is 3 inches. We can use the CDF of the normal distribution to find the probability of someone being less than or equal to 70 inches tall.

To find the probability, we first need to standardize the value using the formula:

z = (x - μ) / σ

where z is the standardized score, x is the value we want to find the probability for, μ is the mean of the distribution, and σ is the standard deviation.

Plugging in the values, we get:

z = (70 - 68) / 3 = 0.67

We can then use a standard normal distribution table or calculator to find the probability associated with this standardized score:

P(Z ≤ 0.67) = 0.7486

So, the probability of someone being less than or equal to 70 inches tall in this population is 0.7486 or about 75%.

We can also use the CDF to find the probability of someone being taller than 70 inches by subtracting the probability of being less than or equal to 70 inches from 1:

P(X > 70) = 1 - P(X ≤ 70)

= 1 - 0.7486

= 0.2514

So, the probability of someone being taller than 70 inches in this population is 0.2514 or about 25%.

In summary, a cumulative distribution function is a function that gives the probability of a random variable being less than or equal to a certain value. It is useful for analyzing and understanding the probabilities of different events and ranges of values in a continuous random variable.

__Mean (μ)__

Mean (μ): The mean is the center point of the distribution. It represents the point around which the data is clustered. The value of the mean determines the location of the peak of the distribution. If the mean is shifted to the right, the entire distribution will shift to the right, and vice versa. For example, if the mean is increased, the peak of the distribution will move to the right, indicating that the data is skewed to the right.

__Standard deviation (σ)__

Standard deviation (σ): The standard deviation is a measure of the spread of the data. It represents how much the data deviates from the mean. A small standard deviation indicates that the data is tightly clustered around the mean, while a large standard deviation indicates that the data is more spread out. The larger the standard deviation, the flatter and wider the curve of the distribution becomes.

Thus, together, the mean and standard deviation of the normal distribution define the shape of the curve. The standard deviation determines the spread of the distribution, while the mean determines the center of the distribution. 

In summary, the mean and standard deviation are important parameters in the normal distribution that determine the center and spread of the distribution, respectively. Changing these parameters will shift and reshape the normal distribution accordingly.

Q5: What is Bernaulli Distribution? Give an Example. What is the difference between Bernoulli
Distribution and Binomial Distribution?

The Bernoulli distribution is a discrete probability distribution that represents the outcomes of a single binary experiment, where there are only two possible outcomes, usually denoted as "success" and "failure".

Using formular:

The Bernoulli distribution is characterized by a single parameter, p, which represents the probability of a success in a single trial. The probability of failure is simply 1-p.

The probability mass function (PMF) of the Bernoulli distribution is:

P(X = k) = p^k * (1-p)^(1-k)

where X is a random variable that takes on the values of 0 or 1 (failure or success), and k can be either 0 or 1.


__Example__

Suppose we have a bag containing 10 marbles, 4 of which are red and 6 of which are green. We randomly select one marble from the bag, and define a success as selecting a red marble. The experiment can only have two possible outcomes: success (selecting a red marble) or failure (selecting a green marble).

The probability of success (p) is the proportion of red marbles in the bag, which is 4/10 or 0.4. The probability of failure (1-p) is the proportion of green marbles in the bag, which is 6/10 or 0.6.

We can model this experiment using the Bernoulli distribution. The random variable X represents the outcome of the experiment, where X=1 if we select a red marble (success), and X=0 if we select a green marble (failure).

The probability of success, P(X=1), is given by the Bernoulli PMF:

P(X=1) = p^1 * (1-p)^(1-1) = p

P(X=0), the probability of failure, is given by:

P(X=0) = p^0 * (1-p)^(1-0) = 1-p

So, the probability distribution of X can be summarized as:

X = 1 with probability p = 0.4
X = 0 with probability 1-p = 0.6

This is an example of a Bernoulli distribution with parameter p=0.4.


The Differences Between Both:

The Bernoulli distribution represents the probability distribution of a single binary experiment, where there are only two possible outcomes, usually denoted as "success" and "failure". The Bernoulli distribution has a single parameter, p, which represents the probability of success in a single trial.

While the binomial distribution, on the other hand, represents the probability distribution of the number of successes in a fixed number of independent Bernoulli trials. In other words, the binomial distribution is the sum of independent and identically distributed Bernoulli random variables. In terms of formula:

The binomial distribution has two parameters: n, the number of trials, and p, the probability of success in a single trial. The probability mass function (PMF) of the binomial distribution is given by:

P(X=k) = (n choose k) * p^k * (1-p)^(n-k)

where X is the random variable representing the number of successes in n trials, k is the number of successes, (n choose k) is the binomial coefficient, which represents the number of ways to choose k items from a set of n items.

On a final note, the key difference between the Bernoulli and binomial distributions is that the Bernoulli distribution represents the probability distribution of a single binary experiment, while the binomial distribution represents the probability distribution of the number of successes in a fixed number of independent Bernoulli trials.

Q8: What is the z score? State the importance of the z score.

The z-score, also known as the standard score, is a statistical measure that indicates how many standard deviations a data point is from the mean of a data set. It is calculated by subtracting the mean from the data point and then dividing the result by the standard deviation. The formula for calculating the z-score is:

z = (x - μ) / σ

where z is the z-score, x is the data point, μ is the mean of the data set, and σ is the standard deviation of the data set.

__Importance__

The z-score is an important statistical measure for several reasons:

__Standardization__

Standardization: The z-score allows for the standardization of data, which makes it possible to compare data from different sources that have different scales and units of measurement.

__Outlier detection__

Outlier detection: The z-score can be used to identify outliers, which are data points that are significantly different from the rest of the data set. Outliers can be important to detect because they can skew the results of statistical analysis.

__Probability Calculation__

Probability calculation: The z-score can be used to calculate the probability of a data point occurring within a certain range of values. This can be useful in determining the likelihood of certain events or outcomes.

__Normal Distribution__
Normal distribution: The z-score is also important in the study of normal distribution, which is a common distribution in statistics. The z-score is used to determine the area under the normal distribution curve, which can be used to calculate probabilities and make predictions.

Q9: What is Central Limit Theorem? State the significance of the Central Limit Theorem.

The Central Limit Theorem is a statistical concept that states that the sample mean of a large number of independent and identically distributed random variables will be approximately normally distributed, regardless of the underlying distribution of the original population.

It simply means that if multiple random samples of the same size from a population is taken, and the mean of each sample is calculated, the distribution of those means will be approximately normal, regardless of the shape of the original population distribution. The mean of the sample means will also be equal to the mean of the original population.

The Central Limit Theorem is important because it allows us to make inferences about a population based on a sample, and it is the foundation of many statistical techniques, such as hypothesis testing and confidence interval estimation.

Some of its assumptions include:

1. The independence of the random variables, 
2. The finite variance of the population
3. The sample size being large enough.

If these assumptions are met, then the Central Limit Theorem can be used to make statistical inferences about a population based on a sample.

Below are the assumptions guiding Central Limit Theorem:


Independence: The random variables in the sample must be independent, meaning that the outcome of one variable does not affect the outcome of the other variables.

Identical distribution: The random variables in the sample must have the same distribution, with the same mean and variance.

Finite variance: The population from which the sample is drawn must have a finite variance.
